```python
# Untuk Mereset harus melakukan impor xlxs Lagi

data_match = data_match.sort_values("Goal Scored", ascending=False)
data_match = data_match.drop(["Team"], axis = 1) # Axis 1 Menghapus line berdasarkan sumbu y ( Kebawah )
data_match.head()
```

```python
data_match["Team"] = data_match["Team"].str.upper()
data_cou["Country"] = data_cou["Country"].str.upper()
new_feature = data_match.merge(data_cou, left_on="Team", right_on="Country", how="left")

# str upper membuat seluruh huruf besar agar dapat di merge
```

```python
new_feature = new_feature.drop("Country", axis = 1)
new_feature["Team"] = new_feature["Team"].str.title()
data_match["Team"] = data_match["Team"].str.title()
data_cou["Country"] = data_cou["Country"].str.title()
new_feature.head()
```

```python
new_feature["Continent"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan matplotlib dalam JupyterBook
%matplotlib inline 
```

```python
data_match.head()
```

```python
# Heatmap
plt.figure(figsize=(20,20))
sns.heatmap(data_match.corr(), annot=True)
plt.show()
```

```python
#Ball Possesion

plt.figure(figsize=(10,10))
sns.scatterplot(x="Passes", y="Ball Possession %", size = "Passes", hue = "Passes", data = data_match)
plt.title("The Influence Of Passes On Ball Possesion", size = 20)
plt.xlabel("Passes")
plt.ylabel("Ball Possesion (%)")
plt.legend()
plt.show()
```

```python
europe_goal = pd.DataFrame(new_feature[new_feature["Continent"]=="Europe"].groupby("Team")["Goal Scored"].sum())
europe_goal = europe_goal.sort_values("Goal Scored", ascending=True) # SORT Goal
europe_goal.head()
```

```python
europe_goal.plot(kind="barh", figsize=(10,8))
# europe_goal.plot(kind="barh", figsize=(10,8)) # Menampilkan bar biasa
plt.title("Goal Scored In Europe Continent", size=20)
plt.ylabel("Country")
plt.xlabel;("Goal Scored")
plt.show()
```

```python
# Box Plot ( Persebaran Data )   
europe_and_america_fouls = pd.DataFrame(new_feature[(new_feature["Continent"] == "Europe") | (new_feature["Continent"] == "South America")].groupby(["Continent","Team"])["Fouls Committed"].sum())
europe_and_america_fouls = europe_and_america_fouls.sort_values("Fouls Committed", ascending=False)
europe_and_america_fouls = europe_and_america_fouls.reset_index()
europe_and_america_fouls.head()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x = europe_and_america_fouls["Continent"], y = europe_and_america_fouls["Fouls Committed"])
plt.title("Fouls Committed (Europe And South America)", size=20)
plt.show()
```
